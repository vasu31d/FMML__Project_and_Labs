## Gradient Descent

```python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x) ## [REQUIRED]
    new_x = previous_x - step_multiplier * gradient ## [REQUIRED]

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))## [REQUIRED])

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

## Gradient Descent

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$

```python
# Challenge: Write a python function mse(y, y_hat) returns the MSE? Call the mse(y, y_hat)
# function and print out the MSE for the y_hat calculated above.

def mse(y, y_hat):
    #mse_calc = 1/7 * sum((y - y_hat)**2)
    #mse_calc = (1/y.size) * sum((y - y_hat)**2)
    mse_calc = np.average((y - y_hat)**2, axis=0)
    return mse_calc
```

## Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

## Define the derivative of h(x)

### Subtask:
Create a Python function `dh(x)` for the derivative of $h(x) = x^4 - 5x^2 - 3x$.


**Reasoning**:
Define the derivative function `dh(x)` as requested in the subtask instructions.



```python
def dh(x):
    return 4 * x**3 - 10 * x - 3
```

## Generate data

### Subtask:
Create an array of x values for plotting.


**Reasoning**:
Create the x values for plotting the function and its derivative.



```python
# Make some data
x_3 = np.linspace(-3, 3, 200)
```

## Plot h(x) and dh(x)

### Subtask:
Plot the function h(x) and its derivative dh(x) side by side.


**Reasoning**:
Plot the function h(x) and its derivative dh(x) side by side as instructed.



```python
# Plot function and derivative side by side
plt.figure(figsize=[15, 5])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)

plt.xlim(-3, 3)
plt.ylim(-10, 20)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('h(x)', fontsize=16)

plt.plot(x_3, h(x_3), color='blue', linewidth=3)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dh(x)', fontsize=16)
plt.grid()
plt.xlim(-3, 3)
plt.ylim(-20, 20)

plt.plot(x_3, dh(x_3), color='skyblue', linewidth=5)

plt.show()
```

## Perform gradient descent

### Subtask:
Apply the `gradient_descent` function to the derivative of h(x) (`dh(x)`) with an initial guess to find a local minimum.


**Reasoning**:
Apply the gradient_descent function to the derivative of h(x) to find a local minimum and print the results.



```python
# Apply gradient descent to find a local minimum of h(x)
local_min_h, list_x_h, deriv_list_h = gradient_descent(derivative_func=dh, initial_guess=0)

# Print the results
print('Local minimum occurs at:', local_min_h)
print('Number of steps:', len(list_x_h))
```

## Visualize gradient descent

### Subtask:
Superimpose the gradient descent steps on the plots of h(x) and dh(x).


**Reasoning**:
Superimpose the gradient descent steps on the plots of h(x) and dh(x) according to the instructions.



```python
# Superimpose the gradient descent calculations on plot

plt.figure(figsize=[15, 5])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)

plt.xlim(-3, 3)
plt.ylim(-10, 20)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('h(x)', fontsize=16)

plt.plot(x_3, h(x_3), color='blue', linewidth=3, alpha=0.8)

plt.scatter(list_x_h, h(np.array(list_x_h)), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dh(x)', fontsize=16)
plt.grid()
plt.xlim(-3, 3)
plt.ylim(-20, 20)

plt.plot(x_3, dh(x_3), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(list_x_h, deriv_list_h, color='red', s=100, alpha=0.5)


plt.show()
```
